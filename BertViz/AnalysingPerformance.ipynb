```python
# Using readlines()
file1 = open('dev.txt', 'r')
Lines = file1.readlines()
```

```python
dev_lines = []
dev_targets = []
dev = {}
dev_y = {}

count = 0
for line in Lines:
  line_parts = line.split()
  if len(line_parts) == 2:
    dev_lines.append(line_parts[0])
    dev_targets.append(line_parts[1])

  else:
    dev[count] = dev_lines
    dev_y[count] = dev_targets
    count = count + 1
    dev_lines = []
    dev_targets = []
```

```python
class_map = {}

for i in range(20):
  class_map[i] ="O"
```

```python
import json
  
# Opening JSON file
f = open('epoch_44.json')
latest = json.load(f)
```

```python
preds = []
targets = []
for i in range(len(latest)):
  targets.append(latest[i]["target"][0:len(dev[i])])
  preds.append(latest[i]["pred"][0:len(dev[i])])
  
  print(targets[len(targets)-1])
  print(dev_y[i])
  print("-----------------------------------------")

  for _, j in enumerate(targets[len(targets)-1]):

    if class_map[j] == "O":
      class_map[j] = dev_y[i][_]
```

```python
class_map
```

```python
per_failed = []
per_succeed = []


loc_failed = []
loc_succeed = []


org_failed = []
org_succeed = []
```

```python
list_of_list_preds =[]
list_of_list_targets =[]
```

```python
for sublist_pred, sublist_target in zip(preds, targets):
  temp_pred = [class_map[pr] for pr in sublist_pred]
  temp_target = [class_map[tr] for tr in sublist_target]

  list_of_list_preds.append(temp_pred)
  list_of_list_targets.append(temp_target)
```

```python
print(list_of_list_preds)
```

```python
print(list_of_list_targets)
```

```python
cnt = 0
for sublist_pred, sublist_target in zip(list_of_list_preds, list_of_list_targets):
  if 'I-PER' in sublist_target:
    inx = sublist_target.index('I-PER')

    if sublist_target[inx] == sublist_pred[inx]:
      per_succeed.append(dev[cnt])

    else:
      per_failed.append(dev[cnt])

  elif 'I-LOC' in sublist_target:
    inx = sublist_target.index('I-LOC')

    if sublist_target[inx] == sublist_pred[inx]:
      loc_succeed.append(dev[cnt])

    else:
      loc_failed.append(dev[cnt])

  elif 'I-ORG' in sublist_target:
    inx = sublist_target.index('I-ORG')

    if sublist_target[inx] == sublist_pred[inx]:
      org_succeed.append(dev[cnt])

    else:
      org_failed.append(dev[cnt])

  cnt = cnt + 1
```

```python
preds_list = [item for sublist in preds for item in sublist]
targets_list = [item for sublist in targets for item in sublist]
```

```python
pred_list_pre = []
target_list_pre = []

#for pr in preds_list:
#    pred_list_pre.append(class_map[pr])

for tr, pr in zip(targets_list, preds_list):
  if class_map[tr]!="O":
    target_list_pre.append(class_map[tr])
    pred_list_pre.append(class_map[pr])
```

```python

# Example of a confusion matrix in Python
from sklearn.metrics import confusion_matrix

results = confusion_matrix(target_list_pre, pred_list_pre, labels=[ "I-PER", "I-LOC", "I-ORG"])#"O",
print(results)
```
